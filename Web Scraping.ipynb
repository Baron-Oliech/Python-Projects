# Web Scraping

```python
from bs4 import BeautifulSoup
import requests
```

```python
url="https://www.altexsoft.com/blog/data-hub/"
```

```python
page=requests.get(url)
```

```python
soup=BeautifulSoup(page.text,'html')
```

```python
print (soup)
```

```python
print (soup.prettify)
```
